Experience Histogram

## Experience Distribution Histogram

```
import matplotlib.pyplot as plt

df["experience_years"].hist(bins=20)
plt.xlabel("Experience Years")
plt.ylabel("Frequency")
plt.title("Experience Distribution")
plt.show()
```

## What is the average skills count?

```
df["skills_count"].mean()
```

## Skills Count Distribution

```
df["skills_count"].hist(bins=10)
plt.xlabel("Skills Count")
plt.ylabel("Frequency")
plt.title("Skills Distribution")
plt.show()
```

## Which job titles are most common?

```
df["job_title"].value_counts().head(10)
```

## Top Job Titles Bar Chart

```
df["job_title"].value_counts().head(10).plot(kind="bar")
plt.xlabel("Job Title")
plt.ylabel("Count")
plt.title("Top Job Titles")
plt.show()
```

## What is the distribution of education levels?

```
df["education_level"].value_counts()
```

## Education Level Pie Chart

```
df["education_level"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.ylabel("")
plt.title("Education Levels")
plt.show()
```

## Which industries have the highest number of employees?

```
df["industry"].value_counts()
```

## Industry Bar Chart

```
df["industry"].value_counts().plot(kind="bar")
plt.xlabel("Industry")
plt.ylabel("Count")
plt.title("Industry Distribution")
plt.show()
```

## Experience vs Skills Scatter Plot

```
plt.scatter(df["experience_years"], df["skills_count"])
plt.xlabel("Experience Years")
plt.ylabel("Skills Count")
plt.title("Experience vs Skills")
plt.show()
```

## Correlation Heatmap

```
import seaborn as sns

sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
